```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

Population_merge_consommation_import_religion_revenues_sans_country_df = pd.read_csv("out.csv")
```

```python
X = Population_merge_consommation_import_religion_revenues_sans_country_df.drop(['import_par_hab',
                                                                                 'consommation_par_hab',
                                                                                 'Code zone',
                                                                                 'Continent',
                                                                                 'Pop'], axis=1)
X
```

```python
from sklearn import datasets
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.cluster import KMeans

transformerVectoriser = ColumnTransformer(
    transformers=[
        ('Encoding zone', OneHotEncoder(sparse=False, handle_unknown='ignore'), ['Zone']),
        ('Encoding religion', OneHotEncoder(sparse=False, handle_unknown='ignore'), ['Religion'])#,
        #('Encoding continent', OneHotEncoder(sparse=False, handle_unknown='ignore'), ['Continent'])
    ],
    remainder='passthrough'
)

X = transformerVectoriser.fit_transform(X)

# scale the features
scaler = StandardScaler()
features = scaler.fit_transform(X)
k_list = range(1, 10)
# run the cluster analysis
sum_of_squared_distances = []
for k in k_list:
    kmeans = KMeans(n_clusters=k)
    kmeans = kmeans.fit(X)
    sum_of_squared_distances.append(kmeans.inertia_)
```

```python
fig, ax = plt.subplots(1,1,figsize=(12,6))
ax.set_ylabel("intertia")
ax.set_xlabel("n_cluster")

ax = plt.plot(k_list, sum_of_squared_distances)
```

```python
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# On instancie notre Kmeans avec 3 clusters : 
kmeans = KMeans(n_clusters=3)
# On l'entraine : 
kmeans.fit(features)
# On peut stocker nos clusters dans une variable labels : 
labels = kmeans.labels_
labels
```

```python
# On peut stocker nos centroids dans une variable : 
from sklearn.cluster import KMeans
centroids = kmeans.cluster_centers_
centroids #étude des centroides
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
```

```python
pca = PCA(n_components=15)
reduced = pca.fit_transform(features)

PC_values = np.arange(pca.n_components) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```
