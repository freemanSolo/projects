On veut faire un test statistique pour verifier l'existance d'une relation entre les revenus par habitants et la consommation de poulet. 
H0 : pas de relation entre les revenus et la consommation de poulet. 
H1 : il y'a une relation entre les revenus et la consommation de poulet. 

```python
from scipy.stats import pearsonr
DB = Population_merge_consommation_import_religion_revenues_sans_country_df.dropna()
Y1=DB['Revenues'].dropna()
Y2=DB['consommation_par_hab'].dropna()

r,p=pearsonr(Y2,Y1)
print(p)
if p<0.05: #alpha value 5%
    print("on rejette l'hypothese 0")
else: print("on accepte l'hypothese 0")
    #il ya donc une relation Revenus/consommation 
```

```python
my_rho = np.corrcoef(Y2,Y1)
print(my_rho)
```
